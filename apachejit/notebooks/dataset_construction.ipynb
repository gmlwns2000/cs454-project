```python
clean = pd.read_csv('../data/clean_filtered.csv').rename(columns={'commit_date': 'date'})
clean_filtered = pd.read_csv('../data/keys_clean_ast.csv')
clean = pd.merge(clean, clean_filtered)
clean
```

```python
dates = fix['date'].tolist()
strata_per_year = 1
strata_size = 60 * 60 * 24 * (365.25 / strata_per_year)
min_date = 1041397200    # Jan 1, 2003
strata = [int((d - min_date) // strata_size) for d in dates]
fix['strata'] = strata
fix
```

```python
dates = buggy['date'].tolist()
strata_per_year = 1
strata_size = 60 * 60 * 24 * (365.25 / strata_per_year)
min_date = 1041397200    # Jan 1, 2003
strata = [int((d - min_date) // strata_size) for d in dates]
buggy['strata'] = strata
buggy
```

```python
dates = clean['date']
strata_per_year = 1
strata_size = 60 * 60 * 24 * (365.25 / strata_per_year)
min_date = 1041397200    # Jan 1, 2003
strata = [int((d - min_date) // strata_size) for d in dates]
clean['strata'] = strata
clean
```

```python
clean['buggy'] = [False for i in range(len(clean))]
clean['fix'] = [False for i in range(len(clean))]
clean = clean[list(buggy.columns.values)]
clean
```

```python
apachejit = pd.concat([buggy, fix, clean])
apachejit['year'] = apachejit['strata'] + 2003
apachejit = apachejit.drop(columns=['strata']).sort_values('year')
apachejit = apachejit[apachejit['year'] <= 2019]
apachejit
```

```python
metrics = pd.read_csv('../data/apache_metrics_kamei.csv')
apachejit = pd.merge(apachejit, metrics).drop(columns=['date'])
apachejit.to_csv('../data/apachejit_total.csv', index=False)
```

```python
b_train = buggy[buggy['strata'] < 14]
f_train = fix[fix['strata'] < 14]
c_train = clean[clean['strata'] < 14]
```

```python
def sample(obj, replace=False):
    return obj.sample(n=min(int(b_train['strata'].value_counts(normalize=True)[obj.name] * (len(b_train) - len(f_train))), len(obj)), replace=replace)

sample_clean = c_train.groupby('strata', as_index=False).apply(sample).reset_index(drop=True)
sample_clean
```
